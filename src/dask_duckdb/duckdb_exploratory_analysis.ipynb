```python
from pathlib import Path
import duckdb
import pandas as pd
from IPython.display import display
pd.options.display.max_columns = None
from dask.dataframe import read_sql_query

cwd = Path.cwd()
path = str(cwd).split("BIG-DATA-PROJECT")
PARQUET_PATH = Path(path[0]) / "BIG-DATA-PROJECT/data/parquet/"
DUCKDB_PATH = Path(path[0]) / "BIG-DATA-PROJECT/data/duckdb/"

# Create a DuckDB connection
conn = duckdb.connect(str(DUCKDB_PATH / "nyc_database.db"))
```

```python
# load the parquet files into conn
conn.execute(f"CREATE TABLE nyc_data AS SELECT * FROM parquet_scan('{PARQUET_PATH / '*.parquet'}')")
```

```python
conn.execute("SHOW TABLES").fetchall()
```

```python
conn.execute("SELECT * FROM nyc_data LIMIT 5").fetch_df()
```

```python
# show all tables
print(conn.execute("SHOW TABLES").fetchall())

# show the schema of the table 'nyc'
conn.execute("DESCRIBE nyc_data").fetchdf()
```

```python
read_sql_query("SELECT * FROM nyc_data LIMIT 5", conn, index_col="Summons Number")
```
